# 2. Cleaning the data

We wil now look at the data itself, and perform any preliminary cleaning that ought to be done before analysis.

```python
autos.describe()
```

Looking at the headline figures, we make a the following oservations:
1. The nr_of_pictures is all 0 , indicating either a data collection error (unlikely that for all listings there are 0 pictures on ebay)
2. Registration year will require investigation, as the minimum registration year is seen as 1000 - unlikely , given that the first car was invented in 1885.
3. There might be missing registration months as well, given that theese can only have values 1-12 , whereas we see a minimum values of 0. It is possible that the nulls are stored as -. 
4. The 'autos.describe' returns the summary statistics of the values stored as floats. Comparing this to 'autos.info' we can see that some of the columns, which we'd expect to be useful to interpret as floats/integeres are stored as strings. These for example, include the 'price' and 'odometer' columns, as these include the currency symbols and distance units


In order to prepare the data for analysis, we will remove the unit descriptors from the individual rows (separating them), and convert the columns into floats. 

We also notice that there are ',' (commas) used in order to seperate the 1000s. We will also remove those to allow for float analysis.

First we convert  the price columns to flot figures

```python
#converting the price column from 'object' to 'string' type
autos['price'] = autos['price'].astype(str)

#applying string operations

autos['price'] = autos['price'].str.replace('$','')
autos['price'] = autos['price'].str.replace(',','')
autos['price'] = autos['price'].astype(int)

#inspecting the data after conversion
autos['price'].describe()
```

Now we convert the odometeer reading column to float values, removing the distance units, and comma 1000s separator. We will also rename the column header from 'odometer' to 'odometer_km' to preserve the information that the values are stored in kilometers

```python
#converting the odometer column from 'object' to 'string' type
autos['odometer'] = autos['odometer'].astype(str)

#applying string operations

autos['odometer'] = autos['odometer'].str.replace('km','')
autos['odometer'] = autos['odometer'].str.replace(',','')
autos['odometer'] = autos['odometer'].astype(int)

#inspecting the data after conversion
autos['odometer'].describe()
```

```python
# 
autos.rename({'odometer' : 'odometer_km'}, axis = 1,  inplace = True)

print(autos.columns)
```

We have now converted the numeric columns to store numeric float values, rather than objects.

# Translating categorical data
In order to make the data more readable for an english-speaking audience, we will translate some of the categorical data into english.



We will do this by the means of mapping dictionaries. The below example translates the german 'ja' , 'nein' into english 'yes' and 'no' by the means of mapping dictionaries.


```python
autos
```

Inspecting the values with object inputs:


```python
print(autos['seller'].value_counts())
print('\n')
print(autos['offer_type'].value_counts())
print('\n')
print(autos['abtest'].value_counts())
print('\n')
print(autos['vehicle_type'].value_counts())
print('\n')
print(autos['gearbox'].value_counts())
print('\n')
print(autos['fuel_type'].value_counts())
```

We find that there are opportunities to translate from german to english. We will again to this by the means of mappings in pandas. 

```python
seller_map = {
    'privat' : 'private'
}

offer_type_map = {
    'Angebot' : 'offer'
}

#'abtest' column is directly translatable so will be omitted

vehicle_type_map = {
    'limousine' : 'limousine',
    'kombi' : 'kombi',
    'kleinwagen' : 'compact',
    'bus' : 'bus',
    'cabrio' : 'convertible',
    'coupe' : 'coupe',
    'suv' : 'suv',
    'andere' : 'other'
}

gearbox_map = {
    'manuell' : 'manual',
    'automatik' : 'automatic'
}

fuel_type_map = {
    'benzin' : 'gasoline',
    'diesel' : 'diesel',
    'lpg' : 'lpg',
    'cng' : 'cng',
    'hybrid' : 'hybrid',
    'elektro' : 'electric',
    'andere' : 'other' 
}
```

```python
#applying the mapping

autos['seller'] = autos['seller'].map(seller_map)
autos['offer_type'] = autos['offer_type'].map(offer_type_map)
autos['vehicle_type'] = autos['vehicle_type'].map(vehicle_type_map)
autos['gearbox'] = autos['gearbox'].map(gearbox_map)
autos['fuel_type'] = autos['fuel_type'].map(fuel_type_map)
```

We note that:

The seperator used between words is an underscore, rather than a space.
The input is user generated, and can be any string of characters.

In order to perform the keyword analysis we will seperate the entries into the list, and identify the most common keywords on the lists. We will also convert all to lower case to make the analysis consistent, as we do not need to  differentiate between capitalised words for this particular analysis. 

Then for each list we create a dicionary that counts the unique words.

In order to do this, we use the 'Counter' library, which implements a simple function for counting unqie values for lists.



```python
from collections import Counter
results = Counter()
autos['name'].str.lower().str.split('_').apply(results.update)
print(results)
```

We can see that the name will most often state the brand and make of the car. The list also counts popular german used to construct the ads.
